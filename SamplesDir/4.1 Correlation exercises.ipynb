```r
library(tidyverse)
options(repr.plot.width=8, repr.plot.height=5)
set.seed(9876)
dia <- diamonds[sample(1:nrow(diamonds), 2500), ]
```

### *Exercise 1*

Make a scatterplot with `carat` as the `x` variable and `price` as the `y` variable. (Hint: use something like `ggplot(data = ...) + geom_point(aes(x = ..., y = ...))`)

- What type of correlation do you see wrt. direction, shape, strength, and outliers?
- What is the correlation coefficient between the two variables?
- Is there any causal relationship between the two variables and if so which one?
